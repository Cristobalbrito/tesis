# Representación de imágenes

En este notebook exploraremos un poco cómo las imágenes se representan en una computadora. Para ello, utilizaremos librerías específicamente diseñadas no solo para el manejo de imágenes, sino de matrices, álgebra lineal y arreglos numéricos multidimensionales como: `numpy`, `opencv`, `matplotlib`, entre otras.

## Importando los recursos

Lo primero que debemos hacer es importar las librerías que necesitaremos.

```python
import numpy as np
import matplotlib.image as mpimg  # Para leer imágenes

import matplotlib.pyplot as plt
import cv2  # OpenCV, una libreria de computer vision

%matplotlib inline
```

## Cargando y mostrando una imagen

```python
# Lee la imagen
image = mpimg.imread('./pug.jpg')

# Imprime las dimensiones:
height, width, channels = image.shape
print("Ancho: {}".format(width))
print("Alto: {}".format(height))
print("Profundidad: {}".format(channels))

# Convierte la imagen a escala de grises
gray_image = cv2.cvtColor(image, cv2.COLOR_RGB2GRAY)

# Muestra imagen en escala de grises
plt.imshow(gray_image, cmap="gray")
print(image)
```

También podemos ver una muestra del arreglo de números que representa a esta imagen:

```python
print(gray_image)
```

Podemos examinar el valor de un píxel específico

```python
x = 300
y = 300
print(gray_image[y, x])
```

Como se trata de un arreglo de números, podemos ejecutar operaciones matemáticas, como hallar el mínimo y el máximo:

```python
max_val = np.amax(gray_image)
min_val = np.amin(gray_image)

print("Máximo: ", max_val)
print("Mínimo: ", min_val)
```

## RGB

Hasta ahora hemos trabajado con una imagen en escala de grises. Sin embargo, es hora de trabajar con colores y, para ello, exploraremos un poco el espacio RGB (Red, Green, Blue | Rojo, Verde, Azul)

Empecemos por cargar una imagen con muchos colores:

```python
colorful_image = mpimg.imread('./colors.jpeg')

plt.imshow(colorful_image)
```

```python
# Aislemos cada canal de color:

red_channel = colorful_image[:, :, 0]
green_channel = colorful_image[:, :, 1]
blue_channel = colorful_image[:, :, 2]

# Ahora, mostremos los tres canales como imágenes en escala de grises. Veremos que cada canal suele enfatizar el color que le corresponde:
f, (ax1, ax2, ax3) = plt.subplots(1, 3, figsize=(20, 10))
ax1.set_title("Canal ROJO")
ax1.imshow(red_channel)

ax2.set_title("Canal VERDE")
ax2.imshow(green_channel)

ax3.set_title("Canal AZUL")
ax3.imshow(blue_channel)
```
